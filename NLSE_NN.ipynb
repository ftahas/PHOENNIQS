Here we construct a Neural Network for learning the solitons solution of the Nonlinear Schroedinger equation (NLSE), given by 
$$
i \partial_t \psi + \frac{1}{2} \nabla^2 \psi + \gamma |\psi|^2 \psi =0
$$
We define the loss function of the artifical network as
$$ 
\mathcal{L} = \left| i \partial_t \psi_\lambda  + \nabla^2 \psi_\lambda - \gamma  |\psi_\lambda|^2 \psi_\lambda \right|^2.
$$


```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# Set device (GPU if available)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Define the neural network (FNN)
class NLSE_Net(nn.Module):
    def __init__(self, layers):
        super(NLSE_Net, self).__init__()
        self.net = nn.Sequential(
            *[nn.Sequential(nn.Linear(layers[i], layers[i+1]), nn.Tanh()) for i in range(len(layers)-2)],
            nn.Linear(layers[-2], layers[-1])
        )

    def forward(self, x, y, t):
        inputs = torch.cat((x, y, t), dim=1)
        return self.net(inputs)

# Define NLSE loss function
def nlse_loss(model, x, y, t):
    psi = model(x, y, t)
    psi_real, psi_imag = psi[:, 0:1], psi[:, 1:2]  # Real & Imaginary parts

    # Compute first-order derivatives using autograd
    psi_real_t = torch.autograd.grad(psi_real, t, grad_outputs=torch.ones_like(psi_real), create_graph=True)[0]
    psi_imag_t = torch.autograd.grad(psi_imag, t, grad_outputs=torch.ones_like(psi_imag), create_graph=True)[0]

    # Compute Laplacian (∆ψ = ψ_xx + ψ_yy)
    psi_real_xx = torch.autograd.grad(torch.autograd.grad(psi_real, x, torch.ones_like(psi_real), create_graph=True)[0],
                                      x, torch.ones_like(psi_real), create_graph=True)[0]
    psi_real_yy = torch.autograd.grad(torch.autograd.grad(psi_real, y, torch.ones_like(psi_real), create_graph=True)[0],
                                      y, torch.ones_like(psi_real), create_graph=True)[0]

    psi_imag_xx = torch.autograd.grad(torch.autograd.grad(psi_imag, x, torch.ones_like(psi_imag), create_graph=True)[0],
                                      x, torch.ones_like(psi_imag), create_graph=True)[0]
    psi_imag_yy = torch.autograd.grad(torch.autograd.grad(psi_imag, y, torch.ones_like(psi_imag), create_graph=True)[0],
                                      y, torch.ones_like(psi_imag), create_graph=True)[0]

    # Compute NLSE residual
    laplacian_real = psi_real_xx + psi_real_yy
    laplacian_imag = psi_imag_xx + psi_imag_yy

    nonlinear_real = (psi_real**2 + psi_imag**2) * psi_real
    nonlinear_imag = (psi_real**2 + psi_imag**2) * psi_imag

    res_real = psi_imag_t + laplacian_real - nonlinear_real
    res_imag = -psi_real_t + laplacian_imag - nonlinear_imag

    loss = torch.mean(res_real**2 + res_imag**2)
    return loss

# Hyperparameters
layers = [3, 64, 64, 64, 2]  # 3 inputs (x, y, t) → hidden layers → 2 outputs (Re(ψ), Im(ψ))
learning_rate = 0.001
epochs = 5000

# Initialize model and optimizer
model = NLSE_Net(layers).to(device)
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# Generate training data (random sample points)
N_train = 10000
x_train = torch.rand(N_train, 1, device=device) * 2 - 1  # [-1,1]
y_train = torch.rand(N_train, 1, device=device) * 2 - 1
t_train = torch.rand(N_train, 1, device=device) * 1       # [0,1]

x_train.requires_grad = True
y_train.requires_grad = True
t_train.requires_grad = True

# Training loop
loss_history = []
for epoch in range(epochs):
    optimizer.zero_grad()
    loss = nlse_loss(model, x_train, y_train, t_train)
    loss.backward()
    optimizer.step()

    loss_history.append(loss.item())
    if epoch % 500 == 0:
        print(f"Epoch {epoch}: Loss = {loss.item():.6f}")

# Plot loss curve
plt.plot(loss_history)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss Curve')
plt.show()
```

# Soliton solution in 2d of NLSE

$$ 
\psi(x,y,t) = A \, \mathrm{sech} \left(B \sqrt{x^2+y^2} \right) e^{i(\omega t + kx)}
$$


```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# Set device (GPU if available)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Define the neural network (FNN)
class NLSE_Net(nn.Module):
    def __init__(self, layers):
        super(NLSE_Net, self).__init__()
        self.net = nn.Sequential(
            *[nn.Sequential(nn.Linear(layers[i], layers[i+1]), nn.Tanh()) for i in range(len(layers)-2)],
            nn.Linear(layers[-2], layers[-1])
        )

    def forward(self, x, y, t):
        inputs = torch.cat((x, y, t), dim=1)
        return self.net(inputs)

# Bright soliton initial condition
def soliton_exact(x, y, t, A=1.0, B=1.0, omega=0.5, k=0.5):
    r = torch.sqrt(x**2 + y**2)
    psi_real = A * (1/torch.cosh(B * r)) * torch.cos(omega * t + k * x)
    psi_imag = A * (1/torch.cosh(B * r)) * torch.sin(omega * t + k * x)
    return psi_real, psi_imag

# Define NLSE loss function
def nlse_loss(model, x, y, t):
    psi = model(x, y, t)
    psi_real, psi_imag = psi[:, 0:1], psi[:, 1:2]

    # Compute derivatives
    psi_real_t = torch.autograd.grad(psi_real, t, torch.ones_like(psi_real), create_graph=True)[0]
    psi_imag_t = torch.autograd.grad(psi_imag, t, torch.ones_like(psi_imag), create_graph=True)[0]

    psi_real_xx = torch.autograd.grad(torch.autograd.grad(psi_real, x, torch.ones_like(psi_real), create_graph=True)[0], x, torch.ones_like(psi_real), create_graph=True)[0]
    psi_real_yy = torch.autograd.grad(torch.autograd.grad(psi_real, y, torch.ones_like(psi_real), create_graph=True)[0], y, torch.ones_like(psi_real), create_graph=True)[0]

    psi_imag_xx = torch.autograd.grad(torch.autograd.grad(psi_imag, x, torch.ones_like(psi_imag), create_graph=True)[0], x, torch.ones_like(psi_imag), create_graph=True)[0]
    psi_imag_yy = torch.autograd.grad(torch.autograd.grad(psi_imag, y, torch.ones_like(psi_imag), create_graph=True)[0], y, torch.ones_like(psi_imag), create_graph=True)[0]

    # Compute NLSE residual
    laplacian_real = psi_real_xx + psi_real_yy
    laplacian_imag = psi_imag_xx + psi_imag_yy

    nonlinear_real = (psi_real**2 + psi_imag**2) * psi_real
    nonlinear_imag = (psi_real**2 + psi_imag**2) * psi_imag

    res_real = psi_imag_t + laplacian_real - nonlinear_real
    res_imag = -psi_real_t + laplacian_imag - nonlinear_imag

    loss = torch.mean(res_real**2 + res_imag**2)
    return loss

# Hyperparameters
layers = [3, 64, 64, 64, 2]
learning_rate = 0.001
epochs = 1000

# Initialize model and optimizer
model = NLSE_Net(layers).to(device)
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# Generate training data (random sample points)
N_train = 10000
x_train = (torch.rand(N_train, 1, device=device) * 2 - 1) * 2  # [-2,2]
y_train = (torch.rand(N_train, 1, device=device) * 2 - 1) * 2
t_train = torch.rand(N_train, 1, device=device) * 1  # [0,1]

x_train.requires_grad = True
y_train.requires_grad = True
t_train.requires_grad = True

# Initialize NN with soliton initial conditions
psi_real_init, psi_imag_init = soliton_exact(x_train, y_train, t_train)
with torch.no_grad():
    model.net[0][0].weight *= 0.01  # Scale initialization

# Training loop
loss_history = []
for epoch in range(epochs):
    optimizer.zero_grad()
    loss = nlse_loss(model, x_train, y_train, t_train)
    loss.backward()
    optimizer.step()

    loss_history.append(loss.item())
    if epoch % 500 == 0:
        print(f"Epoch {epoch}: Loss = {loss.item():.6f}")

# Plot loss curve
plt.plot(loss_history)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss Curve for 2D Soliton Solution')
plt.show()

# Test trained model
x_test = torch.linspace(-2, 2, 100, device=device).view(-1, 1)
y_test = torch.linspace(-2, 2, 100, device=device).view(-1, 1)
t_test = torch.zeros_like(x_test)  # t = 0

with torch.no_grad():
    psi_pred = model(x_test, y_test, t_test)

psi_real_pred = psi_pred[:, 0].cpu().numpy()
psi_imag_pred = psi_pred[:, 1].cpu().numpy()

# Plot predicted soliton solution
plt.figure(figsize=(8, 6))
plt.contourf(x_test.cpu().numpy(), y_test.cpu().numpy(), psi_real_pred.reshape(100, 100), cmap='viridis')
plt.colorbar(label="Re(ψ)")
plt.title("Predicted Soliton Solution (Real Part)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

# Soliton solution in 3d

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# Set device (GPU if available)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Define the neural network (FNN)
class NLSE_Net(nn.Module):
    def __init__(self, layers):
        super(NLSE_Net, self).__init__()
        self.net = nn.Sequential(
            *[nn.Sequential(nn.Linear(layers[i], layers[i+1]), nn.Tanh()) for i in range(len(layers)-2)],
            nn.Linear(layers[-2], layers[-1])
        )

    def forward(self, x, y, z, t):
        inputs = torch.cat((x, y, z, t), dim=1)
        return self.net(inputs)

# Bright soliton initial condition
def soliton_exact(x, y, z, t, A=1.0, B=1.0, omega=0.5, k=0.5):
    r = torch.sqrt(x**2 + y**2 + z**2)
    psi_real = A * (1/torch.cosh(B * r)) * torch.cos(omega * t + k * x)
    psi_imag = A * (1/torch.cosh(B * r)) * torch.sin(omega * t + k * x)
    return psi_real, psi_imag

# Define NLSE loss function
def nlse_loss(model, x, y, z, t):
    psi = model(x, y, z, t)
    psi_real, psi_imag = psi[:, 0:1], psi[:, 1:2]

    # Compute derivatives
    psi_real_t = torch.autograd.grad(psi_real, t, torch.ones_like(psi_real), create_graph=True)[0]
    psi_imag_t = torch.autograd.grad(psi_imag, t, torch.ones_like(psi_imag), create_graph=True)[0]

    psi_real_xx = torch.autograd.grad(torch.autograd.grad(psi_real, x, torch.ones_like(psi_real), create_graph=True)[0], x, torch.ones_like(psi_real), create_graph=True)[0]
    psi_real_yy = torch.autograd.grad(torch.autograd.grad(psi_real, y, torch.ones_like(psi_real), create_graph=True)[0], y, torch.ones_like(psi_real), create_graph=True)[0]
    psi_real_zz = torch.autograd.grad(torch.autograd.grad(psi_real, z, torch.ones_like(psi_real), create_graph=True)[0], z, torch.ones_like(psi_real), create_graph=True)[0]

    psi_imag_xx = torch.autograd.grad(torch.autograd.grad(psi_imag, x, torch.ones_like(psi_imag), create_graph=True)[0], x, torch.ones_like(psi_imag), create_graph=True)[0]
    psi_imag_yy = torch.autograd.grad(torch.autograd.grad(psi_imag, y, torch.ones_like(psi_imag), create_graph=True)[0], y, torch.ones_like(psi_imag), create_graph=True)[0]
    psi_imag_zz = torch.autograd.grad(torch.autograd.grad(psi_imag, z, torch.ones_like(psi_imag), create_graph=True)[0], z, torch.ones_like(psi_imag), create_graph=True)[0]

    # Compute NLSE residual
    laplacian_real = psi_real_xx + psi_real_yy + psi_real_zz
    laplacian_imag = psi_imag_xx + psi_imag_yy + psi_imag_zz

    nonlinear_real = (psi_real**2 + psi_imag**2) * psi_real
    nonlinear_imag = (psi_real**2 + psi_imag**2) * psi_imag

    res_real = psi_imag_t + laplacian_real - nonlinear_real
    res_imag = -psi_real_t + laplacian_imag - nonlinear_imag

    loss = torch.mean(res_real**2 + res_imag**2)
    return loss

# Hyperparameters
layers = [4, 64, 64, 64, 2]
learning_rate = 0.001
epochs = 1000

# Initialize model and optimizer
model = NLSE_Net(layers).to(device)
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# Generate training data (random sample points)
N_train = 20000
x_train = (torch.rand(N_train, 1, device=device) * 2 - 1) * 2  # [-2,2]
y_train = (torch.rand(N_train, 1, device=device) * 2 - 1) * 2
z_train = (torch.rand(N_train, 1, device=device) * 2 - 1) * 2
t_train = torch.rand(N_train, 1, device=device) * 1  # [0,1]

x_train.requires_grad = True
y_train.requires_grad = True
z_train.requires_grad = True
t_train.requires_grad = True

# Initialize NN with soliton initial conditions
psi_real_init, psi_imag_init = soliton_exact(x_train, y_train, z_train, t_train)
with torch.no_grad():
    model.net[0][0].weight *= 0.01  # Scale initialization

# Training loop
loss_history = []
for epoch in range(epochs):
    optimizer.zero_grad()
    loss = nlse_loss(model, x_train, y_train, z_train, t_train)
    loss.backward()
    optimizer.step()

    loss_history.append(loss.item())
    if epoch % 500 == 0:
        print(f"Epoch {epoch}: Loss = {loss.item():.6f}")

# Plot loss curve
plt.plot(loss_history)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss Curve for 3D Soliton Solution')
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the PINN network
class PINN_NLSE(nn.Module):
    def __init__(self, input_dim=4, hidden_dim=128, layers=5):
        super(PINN_NLSE, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.Tanh(),
            *[nn.Sequential(nn.Linear(hidden_dim, hidden_dim), nn.Tanh()) for _ in range(layers - 2)],
            nn.Linear(hidden_dim, 2)  # Output two values: real and imaginary parts of psi
        )
    
    def forward(self, x):
        return self.net(x)

# Define the loss function for the NLSE
class NLSE_Loss(nn.Module):
    def __init__(self, g=-1.0):  # g < 0 for attractive interactions
        super(NLSE_Loss, self).__init__()
        self.g = g
    
    def forward(self, model, inputs):
        inputs.requires_grad_()
        output = model(inputs)
        psi_real, psi_imag = output[:, 0], output[:, 1]
        
        # Compute time derivatives
        psi_t_real = torch.autograd.grad(psi_real, inputs, grad_outputs=torch.ones_like(psi_real), create_graph=True, allow_unused=True)[0]
        psi_t_imag = torch.autograd.grad(psi_imag, inputs, grad_outputs=torch.ones_like(psi_imag), create_graph=True, allow_unused=True)[0]
        
        if psi_t_real is None or psi_t_imag is None:
            return torch.tensor(0.0, requires_grad=True, device=inputs.device)  # Prevent NoneType error
        
        psi_t_real, psi_t_imag = psi_t_real[:, 3], psi_t_imag[:, 3]
        
        # Compute Laplacian
        laplacian_real, laplacian_imag = 0, 0
        for i in range(3):
            grad_real = torch.autograd.grad(psi_real, inputs[:, i], grad_outputs=torch.ones_like(psi_real), create_graph=True, allow_unused=True)[0]
            grad_imag = torch.autograd.grad(psi_imag, inputs[:, i], grad_outputs=torch.ones_like(psi_imag), create_graph=True, allow_unused=True)[0]
            
            if grad_real is not None:
                laplacian_real += torch.autograd.grad(grad_real, inputs[:, i], grad_outputs=torch.ones_like(grad_real), create_graph=True, allow_unused=True)[0] or torch.zeros_like(grad_real)
            
            if grad_imag is not None:
                laplacian_imag += torch.autograd.grad(grad_imag, inputs[:, i], grad_outputs=torch.ones_like(grad_imag), create_graph=True, allow_unused=True)[0] or torch.zeros_like(grad_imag)
        
        # Compute NLSE residuals
        residual_real = psi_t_real + 0.5 * laplacian_real + self.g * (psi_real**2 + psi_imag**2) * psi_real
        residual_imag = psi_t_imag + 0.5 * laplacian_imag + self.g * (psi_real**2 + psi_imag**2) * psi_imag
        
        loss = torch.mean(residual_real**2 + residual_imag**2)
        return loss

# Training the PINN
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = PINN_NLSE().to(device)
loss_fn = NLSE_Loss().to(device)
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Generate training data
num_samples = 10000
x = torch.rand(num_samples, 1, device=device) * 2 - 1  # [-1, 1]
y = torch.rand(num_samples, 1, device=device) * 2 - 1
z = torch.rand(num_samples, 1, device=device) * 2 - 1
t = torch.rand(num_samples, 1, device=device) * 2 - 1
inputs = torch.cat((x, y, z, t), dim=1).requires_grad_()

# Training loop
num_epochs = 5000
for epoch in range(num_epochs):
    optimizer.zero_grad()
    loss = loss_fn(model, inputs)
    loss.backward()
    optimizer.step()
    
    if epoch % 500 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.6f}")

# Save trained model
torch.save(model.state_dict(), "pinn_nlse_3d.pth")
print("Training Complete!")
```

```python
# Load and visualise
def plot_soliton():
    model.load_state_dict(torch.load("pinn_nlse_3d.pth"))
    model.eval()
    
    x_vals = np.linspace(-1, 1, 50)
    y_vals = np.linspace(-1, 1, 50)
    z_vals = np.linspace(-1, 1, 50)
    
    X, Y, Z = np.meshgrid(x_vals, y_vals, z_vals, indexing='ij')
    coords = torch.tensor(np.column_stack([X.ravel(), Y.ravel(), Z.ravel(), np.full_like(X.ravel(), 0.0)]), dtype=torch.float32, device=device)
    
    with torch.no_grad():
        psi_pred = model(coords)
        psi_real, psi_imag = psi_pred[:, 0].cpu().numpy(), psi_pred[:, 1].cpu().numpy()
        psi_magnitude = np.sqrt(psi_real**2 + psi_imag**2)
    
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(X.ravel(), Y.ravel(), Z.ravel(), c=psi_magnitude, cmap='viridis', marker='o')
    ax.set_title("Soliton Solution |ψ(x, y, z, t=0)|²")
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    plt.show()

# Run the visualisation
plot_soliton()

print("Visualisation Complete!")
```
